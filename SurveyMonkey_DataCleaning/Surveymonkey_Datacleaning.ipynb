```python
import pandas as pd
```

Importing Dataset

```python
df = pd.read_excel("/content/Data - Survey Monkey Output.xlsx",sheet_name="Edited_Data")
df
```

Created a copy of our dataset incase we need to revert back

```python
df_modified = df.copy()
df_modified
```

Dropping columns we dont need

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1']
columns_to_drop
```

```python
df_modified.drop(columns = columns_to_drop,inplace = True)
```

```python
df_modified
```

Unpivoting our dataframe


```python
id_vars = list(df_modified.columns)[0 : 8]
id_vars
```

```python
value_vars = list(df_modified.columns)[8 : ]
#value_vars
```

```python
df_melted = pd.melt(df_modified,id_vars=id_vars,value_vars = value_vars,var_name ="Question + Subquestion" ,value_name = "Answer")
df_melted
```

```python
df_melted.columns
```

```python
df_melted_2 = df_melted.copy()
df_melted_2
```

Created a Seperate column which only contains Questions

```python
df_melted_2[["Question", "to_drop"]] = df_melted_2["Question + Subquestion"].str.split(" - ", expand=True)
```

```python
df_melted_2.drop(columns = "to_drop",inplace = True)
df_melted_2
```

Finding number of people responded for a question

```python
respondents = df_melted_2[df_melted_2["Answer"].notna()]
respondents = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
```

```python
respondents.rename(columns={"Respondent ID":"Respondents"},inplace = True)
respondents
```

```python
df_merged = pd.merge(df_melted_2,respondents,how="left" ,left_on="Question" , right_on="Question" ,)
print("Before",len(df_melted_2))
print("After",len(df_merged))
df_merged
```

Finding Number of people gave same answer per question

```python
same_answer = df_melted_2 # [df_melted_2["Answer"].notna()]
same_answer = same_answer.groupby(["Question + Subquestion","Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID":"Same Answer"},inplace = True)
same_answer
```

```python
df_merged_2 = pd.merge(df_merged,same_answer,how="left" ,left_on=["Question + Subquestion","Answer"] , right_on=["Question + Subquestion","Answer"] ,)
print("Before",len(df_merged))
print("After",len(df_merged_2))
df_merged_2
```

Filling Nan with 0 in Same Answer Cloumn

```python
df_merged_2["Same Answer"].fillna(0,inplace = True)
df_merged_2["Answer"].fillna("Not Answered",inplace = True)
df_merged_2
```

```python
df_merged_2.columns
```

Renaming column Headers

```python
df_cleaned_1 = df_merged_2.copy()
df_cleaned_1.rename(columns = {'Identify which division you work in. - Response':"Division Primary",
       'Identify which division you work in. - Other (please specify)':"Division Secondary",
       'Which of the following best describes your position level? - Response':"Position",
       'Which generation are you apart of? - Response':"Generation",
       'Please select the gender in which you identify. - Response':"Gender",
       'Which duration range best aligns with your tenure at your company? - Response':"Tenure",
       'Which of the following best describes your employment type? - Response':"Employment Type"},inplace = True)
df_cleaned_1
```

```python
df_cleaned_1.columns
```

Combining Divsion primary and Division Secondary Columns

```python
df_cleaned_2 = df_cleaned_1.copy()
df_cleaned_2['Division Primary'] = df_cleaned_2.apply(lambda row: row['Division Secondary'] if pd.notna(row['Division Primary']) and row['Division Primary'] == 'Other (please specify)' else row['Division Primary'], axis=1)
df_cleaned_2
```
